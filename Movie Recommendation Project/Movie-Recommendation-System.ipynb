**Creating the rec system**

```python
movie_matrix=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
movie_matrix
```

```python
movie_matrix.shape
```

```python
movie_matrix.columns
```

considering star wars first

```python
starwars_user_rating=movie_matrix['Star Wars (1977)']
```

```python
starwars_user_rating
```

```python
similiar_to_starwars=movie_matrix.corrwith(starwars_user_rating)
similiar_to_starwars
```

```python
corr_starwars=pd.DataFrame(similiar_to_starwars,columns=['correlation'])
corr_starwars
```

```python
# we dont want the NaN values, so to drop them:
corr_starwars.dropna(inplace=True)
corr_starwars
```

```python
desc_corr_starwars=corr_starwars.sort_values('correlation',ascending=False)
desc_corr_starwars.head(50)
```

```python
ratings
```

```python
merged_desc_corr_starwars=pd.merge(desc_corr_starwars,ratings,on="title")
```

```python
merged_desc_corr_starwars
```

```python
# the corr for hollow reed ans starwars is 1.0 but no.of is just 6 and we cant base our rec system on such small data. 
# hence, we set a threshold on the no.of ratings for that movie to be considered in the rec system, by merging ratings and corr_starwars data
# can also use below for a different version of merged data that doesnt include avg rating
merged_desc_corr_starwars=desc_corr_starwars.join(ratings['num_of_ratings'])
```

```python
merged_desc_corr_starwars
```

```python
filtered_desc_corr_starwars=merged_desc_corr_starwars[merged_desc_corr_starwars['num_of_ratings']>100]
```

Generalising for all movies

```python
def predict_movies(movie_name):
    movie_ratings=movie_matrix[movie_name]
    #print(movie_ratings)
    similar_movies=movie_matrix.corrwith(movie_ratings)
    similar_movies.dropna(inplace=True)
    #print(similar_movies)
    similar_movies=pd.DataFrame(similar_movies,columns=["correlation"])
    #print(similar_movies)
    merged_similar_movies=similar_movies.join(ratings['num_of_ratings'])
    #print(merged_similar_movies)
    final_similar_movies=merged_similar_movies[merged_similar_movies['num_of_ratings']>100].sort_values('correlation',ascending=False)
    #print(final_similar_movies)
    final_similar_movies=final_similar_movies[final_similar_movies['correlation']>0]
    
    return final_similar_movies
```
